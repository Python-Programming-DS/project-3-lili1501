```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

##### Load and Read Data

```python
df = pd.read_csv("new_train_EGN5442.csv")
df.head()
```

```python
df.info()
```

```python
df.describe(include='all')
```

##### Remove any unwanted characters present under any column

```python
df = df.replace(r'[^A-Za-z0-9\.\-]', '', regex=True)
df.head()
```

##### Further removing unwanted columns such Unnamed: or any fully empty cols that are present

```python
df = df.drop(columns=['Unnamed: 0','x30'])
df.head()
```

```python
before = df.shape
print(before)
```

##### Removing duplicate rows if any

```python
df = df.drop_duplicates()
df.shape
```

##### To handle missing values in some rows

##### if it's numeric column fill nan values with median

```python
for col in df.select_dtypes(include=['float','int']).columns:
    df[col] = df[col].fillna(df[col].median())
```

##### if it's categorical column fill nan values with mode

```python
for col in df.select_dtypes(include=['object']).columns:
    df[col] = df[col].fillna(df[col].mode()[0])
```

##### Looking at the object cols from info

```python
df[['x2','x3','x14','x25','x26','x29']].head()
```

##### Using one hot encoding label for categorical column x3, x25, x26 

```python
df = pd.get_dummies(df, columns=['x3', 'x25', 'x26'], drop_first=True)
df.head()
```

##### Converting the numeric looking col to numeric

```python
df['x2'] = pd.to_numeric(df['x2'], errors='coerce')
df['x29'] = pd.to_numeric(df['x29'], errors='coerce')
df['x14'] = pd.to_numeric(df['x14'], errors='coerce')
df.head()
```

##### To check for highly correlated pairs

```python
corr_matrix = df.corr().abs()
high_corr_pairs = np.where(corr_matrix > 0.8)

high_corr = [
    (df.columns[i], df.columns[j], corr_matrix.iloc[i, j])
    for i, j in zip(*high_corr_pairs)
    if i != j and i < j
]

high_corr
```

##### Checking x5,x6,x23 and x24

```python
df[['x5','x6','x23','x24']]
```

##### x5 and x6 are have perfect correlation, they both are similar. One can be dropped. Similary on an overall level both x23 and x24 have high correlation and one can be dropped

```python
df.drop(columns=['x6','x24'],inplace=True)
df.head()
```

##### Identifying numeric cols

```python
num_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
num_cols
```

##### Filtering continuous cols

```python
continuous_cols = [col for col in num_cols if df[col].nunique() > 10]
continuous_cols
```

```python
df.nunique()
```

##### We have total 103904 rows and x2 has 103904 as id col,this feature won't give us much meaningful info while modeling so will drop x2

```python
df.drop(columns=['x2'],inplace=True)
df.head()
```

#### Binning the continuos cols to qualtile to understand how y changes across ranges

```python
qbin_cols =['x5','x23', 'x29', 'x31']
df[qbin_cols].isna().sum()
```

```python
np.isinf(df[['x5','x23','x29','x31']]).sum()
```

```python
df['x31'] = df['x31'].replace([np.inf, -np.inf], np.nan)
df['x31'] = df['x31'].fillna(df['x31'].median())
```

```python
for col in qbin_cols:
    df[col + '_qbin'] = pd.qcut(df[col].rank(method='first'), q=4,labels=[1, 2, 3, 4],duplicates='drop')
    df[col + "_qbin"] = df[col + "_qbin"].astype(int)
df.head()
```

```python
df.isna().sum()[df.isna().sum() > 0]
```

```python
df['x14'] = df['x14'].fillna(df['x14'].median())
```

```python
df.isna().sum()[df.isna().sum() > 0]
```

```python
sns.countplot(x=df['y'])
plt.title("Target Distribution")
plt.show()
```

#### Visulaize any null values through heat map

```python
plt.figure(figsize=(12,6))
sns.heatmap(df.isnull(), cbar=False)
```

##### Visulaize the continuous column variable for analysis

```python
cols = ['x5','x23','x29','x31']

plt.figure(figsize=(12, 10))
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

for ax, col in zip(axes.flatten(), cols):
    sns.histplot(df[col], bins=30, kde=True, ax=ax)
    ax.set_title(f"Distribution of {col}")

plt.tight_layout()
plt.show()
```

```python
cols = ['x5','x23','x29','x31']

plt.figure(figsize=(12, 10))
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
for ax, col in zip(axes.flatten(), cols):
    sns.barplot(x=col + '_qbin', y='y', data=df, ax= ax)
    ax.set_title(f"Average y Across {col} Bins")
    # ax.tick_params(axis='x', rotation=30)

plt.tight_layout()
plt.show()
```
